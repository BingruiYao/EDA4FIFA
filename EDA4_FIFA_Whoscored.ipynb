**5.1 Normality test for 'Best Overall Rating' in FIFA（shapiro）**

```python
from scipy import stats
stat, p = stats.shapiro(fifa17_df['Best Overall Rating'])
print(f"Statistic: {stat}, p-value: {p}")

if p > 0.05:
    print("Data follows a normal distribution")
else:
    print("Data do not follow a normal distribution")
```

**5.2 Normality test for 'Best Overall Rating' in FIFA（anderson）**

```python
from scipy.stats import anderson
result = anderson(fifa22_df['Best Overall Rating'], dist='norm')
print(f"Statistic: {result.statistic}, Critical Values: {result.critical_values}")

if result.statistic < result.critical_values[2]:  # usually 0.05
    print("Data follows a normal distribution")
else:
    print("Data do not follow a normal distribution")
```

**4.1 Box Plot of Ratings by Classification in FIFA17**

```python
import matplotlib.pyplot as plt

# Create the box plot
plt.figure(figsize=(8, 6))
fifa17_dfr.boxplot(column='Best.Overall.Rating', by='Classification', grid=False)
plt.title('Box Plot of Ratings by Classification in FIFA17')
plt.suptitle('')  # Remove the automatic title
plt.xlabel('Classification')
plt.ylabel('Rating')

plt.show()
```

**4.2 Box Plot of Ratings by Classification across Years (FIFA17-FIFA22)**

```python
import seaborn as sns

fifa17_dfr['Year'] = 'FIFA17'
fifa18_dfr['Year'] = 'FIFA18'
fifa19_dfr['Year'] = 'FIFA19'
fifa20_dfr['Year'] = 'FIFA20'
fifa21_dfr['Year'] = 'FIFA21'
fifa22_dfr['Year'] = 'FIFA22'

# Concatenate all DataFrames into one
df_all_years = pd.concat([fifa17_dfr,fifa18_dfr,fifa19_dfr,fifa20_dfr,fifa21_dfr,fifa22_dfr])

# Plotting using seaborn
plt.figure(figsize=(14, 8))
plt.xticks(fontsize=18)
sns.boxplot(x='Year', y='Best.Overall.Rating', hue='Classification', data=df_all_years)

plt.title('Box Plot of Ratings by Classification across Years (FIFA17-FIFA22)',fontsize=20)
plt.xlabel('Year',fontsize=18)
plt.ylabel('Best Overall Rating',fontsize=18)
plt.legend(title='Classification',fontsize=16)
plt.savefig(r'D:\Project4NUS\Visualization_FIFA\boxplot_classification.png')
plt.show()
```

**4.3 Box Plot of goalkeeper attributes grouped by whether player is goalkeeper in FIFA each year**

**4.3.1 FIFA17**

```python
#FIFA17
import numpy as np
fifa_all['GK_or_not'] = np.where(fifa_all['Position'].str.endswith('GK', na=False), 'GK', 'NGK')
fifa17_dfr=fifa_all[fifa_all['Season']==2017]
variables = ['GKDiving', 'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes']
group_col = 'GK_or_not'

# Transform dataframe to fit seaborn figure construction format
df_melted = pd.melt(fifa17_dfr, id_vars=[group_col], value_vars=variables, 
                    var_name='Attribute', value_name='Score')

plt.figure(figsize=(14, 8))
plt.xticks(fontsize=16) 
sns.boxplot(x='Attribute', y='Score', hue=group_col, data=df_melted)

plt.title('Boxplot of goalkeeper attributes grouped by whether player is goalkeeper in FIFA17',fontsize=20) 
plt.xlabel("Goalkeepers' Attributes",fontsize=18 )
plt.ylabel('Score',fontsize=18 )
plt.legend(title='Group',fontsize=16) 
plt.savefig(r'.\Visualization_FIFA\boxplot_GK_or_not_17.png')
plt.show()
```
